```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# 1. Pedir upload da imagem
uploaded = files.upload()

# Pegar o nome do arquivo enviado
nome_arquivo = list(uploaded.keys())[0]

# 2. Carregar a imagem
imagem_colorida = cv2.imread(nome_arquivo)

# Converter para RGB para exibir corretamente
imagem_rgb = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)

# 3. Converter para tons de cinza (0-255)
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)

# 4. Binarizar (0 e 255)
_, imagem_binaria = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)

# Mostrar resultados
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Colorida")
plt.imshow(imagem_rgb)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Cinza (0-255)")
plt.imshow(imagem_cinza, cmap="gray")
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Binarizada (0/255)")
plt.imshow(imagem_binaria, cmap="gray")
plt.axis("off")

plt.show()
```
